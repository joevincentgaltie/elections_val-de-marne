### Introduction

```python
import pandas as pd 
import folium 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_csv('fichierelections.csv',low_memory=False)
```

```python
df=df.dropna(axis=0, thresh=7)
df['TOUR']=df['TOUR'].fillna(2)
```

```python
df['LIBCOM']=df['LIBCOM'].replace({  'Boissy-Saint-LÃ©ger':'Boissy-Saint-Léger', 'ChenneviÃ¨res-sur-Marne':'Chennevières-sur-Marne', 'CrÃ©teil':'Créteil',
       'Le Kremlin-BicÃªtre':'Le Kremlin-Bicêtre', 'Limeil-BrÃ©vannes':'Limeil-Brévannes', 'PÃ©rigny-sur-Yerres':'Périgny-sur-Yerres',
       'Le Plessis-TrÃ©vise':'Le Plessis-Trévise', 'Saint-MandÃ©':'Saint-Mandé', 'Saint-Maur-des-FossÃ©s':'Saint-Maur-des-Fossés'})
df['LIBCOM'].unique()   
```

```python
df.info()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
df['Nua'].unique()
```

```python
df['Nua']=df['Nua'].replace({'LLR':'LR', 'LUG':'UG', 'LEXG':'EXG', 'LCOM':'COM', 'LDVD':'DVD', 'LREM':'LREM', 'LDVC':'DVC', 'LVEC':'VEC',
       'LDVG':'DVG', 'LDIV':'DIV', 'LNC':'NC', 'LUC':'UC', 'LFI':'LFI', 'LUDI':'UDI', 'LECO':'ECO', 'LMDM':'MODEM', 'LUD':'UD',
       'LRN':'RN', 'LSOC':'SOC', 'FI':'LFI', 'REM':'LREM', 'MDM':'MODEM', 'FN':'RN', 'LFG':'LFI', 'LDLF':'DLF', 'LFN':'RN', 'BC-DVG':'DVG', 'BC-FN':'RN', 'BC-UD':'UD',
       'BC-UG':'UG', 'BC-FG':'FG', 'BC-DIV':'DIV', 'BC-PG':'PG', 'BC-VEC':'VEC', 'BC-SOC':'SOC', 'BC-DLF':'DLF',
       'BC-DVD':'DVD'})
df['Nua'].unique()
```

```python
df['%Exp/Ins']=df['%Exp/Ins'].str.replace(',','.')
df['%Exp/Ins']
df=df.astype({'%Exp/Ins':float})
df['CODCAN']=df['CODCAN'].fillna(0)
df['CODCAN']=df['CODCAN']+94000
df=df.astype({'CODCAN':int})
```

```python
df=df.astype({'CODCAN':str})
df=df.drop_duplicates()
```

#### Municipales 2020

```python
#On charge la carte
map_mun = folium.Map(location=[48.780927,2.4699193],zoom_start=11.2)

tooltip = folium.GeoJsonTooltip(
    fields=['code'],
    aliases=['Canton'],
    localize=True,
    sticky=True,
    labels=True,
    style="""
        background-color: #F0EFEF;
        border: 2px solid black;
        border-radius: 3px;
        box-shadow: 3px;
    """,
    max_width=800,
)
folium.GeoJson(
    "94.json",
    name='geojson',
    tooltip=tooltip,
).add_to(map_mun)
```

```python
map_mun
```

```python
#1er tour
df1=df[(df['Election']=='Municipales2020') & (df['Nua'].isin(centre_droit_elargi)) & (df['TOUR']==1)]
df2= df[(df['Election']=='Municipales2020') & (df['Nua'].isin(centre_droit_elargi)) & (df['TOUR']==1)].groupby(['CODCAN','CODBUR'],as_index=False).sum()
df2=df2.drop(['CODCIR','%Exp/Ins','TOUR','CODBUR','Exprimes','Votants','Inscrits','Nuls'],axis=1)


a=df[df['Election']=='Municipales2020'][['CODCAN','CODBUR','Exprimes','Votants','Inscrits','Nuls']].drop_duplicates()
b=a.groupby('CODCAN',as_index=False).sum()
b=b.drop('CODBUR',axis=1)
df2=df2.groupby('CODCAN',as_index=False).sum()

df2=df2.merge(b,how='left')
df2.loc[:,'%Voix_canton']=round(100*df2['Voix']/df2['Exprimes'],2)


df2


# print(df1['LIBCOM'].unique())
# # print(df['LIBCOM'].unique())

# df[(df['LIBCOM']=='Villeneuve-le-Roi')&(df['Election']=='Municipales2020')&(df['Nom']=='GONZALES')]
# # &(df['Election']=='Municipales2020')&(df['Nom']=='Gonzalès')
```

```python
choropleth= folium.Choropleth(
    geo_data="94.json",
    name='chloropeth',
    data=df2,
    columns=['CODCAN','%Voix_canton'],
    key_on='properties.code',
    fill_color='BuPu',
    fill_opacity=1,
    line_opacity=0.3,
    legend_name='Resultats de l udi aux municipales en %',
    highlight =True,
).add_to(map_mun)
    
folium.LayerControl().add_to(map_mun)
map_mun

map_mun.save('map_mun.html')
```
