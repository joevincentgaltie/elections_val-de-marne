## Départementales 2015 


```python
departementales=pd.read_csv('DP15_Bvot_T1T2.csv',delimiter=";")
```

```python
departementales
```

```python
departementales94=departementales[departementales['CODDPT']=='94']
departementales94.head(100)

```

```python
departementales94.loc[:,'%EXP']=departementales94['NBRVOIX']/departementales94['NBREXP']
departementales94.loc[:,'%ABS']=(departementales['NBRINS']-departementales94['NBREXP'])/departementales94['NBRINS']
departementales94['CODCAN']=94000+departementales94['CODCAN']
dept94 = departementales94
dept94
```

```python
dept94['CODNUA'].unique()
```

```python
dept94agre=dept94.groupby(['CODCAN','LIBSUBCOM','NUMTOUR','CODNUA']).mean()
dept94agre.head(400)

```

```python
dept94['CODCAN']
```

```python
dept94ud=dept94[(dept94['CODNUA']=='BC-UD')& (dept94['NUMTOUR']==1)]
dept94ud=dept94ud.groupby(['CODCAN'],as_index=False).mean()
dept94ud['CODCAN']=dept94ud['CODCAN'].astype(str)
dept94ud
```

### Visualisation 



```python
import folium 
```

```python
map = folium.Map(location=[48.780927,2.4699193],zoom_start=11.2)
map
```

```python
folium.GeoJson(
    "94.json",
    name='geojson'
).add_to(map)
```

```python
map
```

```python
folium.Choropleth(
    geo_data="94.json",
    name='choropleth',
    data=dept94ud,
    columns=['CODCAN', '%EXP'],
    key_on='properties.code',
    fill_color='BuPu',
    fill_opacity=0.7,
    line_opacity=0.3,
    legend_name='Resultats de l Union de la Droite aux Departementales de 2015  (en %)',
    highlight =False ,
    Reset= True
).add_to(map)

map
```

```python
map.save('résultats1ertour.html')
```

```python
folium.LayerControl().add_to(map)
map
```
